# Simulations for determining best values of $\lambda$

```python
# TEST

import numpy
import os

import helpers
import matplotlib

import astropy
from astropy.io import fits

cmap = 'turbo'
matplotlib.rcParams['figure.figsize'] = [10, 5]
```

## Determine number of sample points. Use exponential curve instead of linear to allow for better coverage for lower values of lambda

```python
num_samples = 50
lambda_low = 0.001
lambda_high = 10

step = 1 / float(num_samples)

x_vals = numpy.arange(0, 1 + step, step)

lambdas = [0] * len(x_vals)
for i, v in enumerate(x_vals):
    lambdas[i] = helpers.exp_growth(v, lambda_low, lambda_high, steepness = 1)
    
helpers.plot1D(range(0, len(lambdas)), lambdas, "Iterations", "$\lambda$")
```

## Test case 1: SGR A* from Meerkat Survey with 249600 visibilities

```python
gt = helpers.readFits("lambda_tests/SGRA_full_gt.fits")
dirty = helpers.readFits("lambda_tests/SGRA_full_dirty.fits")
psf = helpers.readFits("lambda_tests/SGRA_full_psf.fits")

helpers.plotGDP(gt, dirty, psf, cmap)
```

```python
snrs = helpers.read_csv("results/lambda_results/SGRA_full/snr.dat")
snrs = [float(x) for x in snrs]
maxidx = numpy.argmax(snrs)
print("Best Lambda: " + str(lambdas[maxidx]))
helpers.plot1D(lambdas, snrs, "$\lambda$", "SNR")

#rmses = helpers.read_csv("results/lambda_results/SGRA_full/rmse.dat")
#rmses = [float(x) for x in rmses]
#helpers.plot1D(lambdas, rmses, "$\lambda$", "RMSE")

#abserr = helpers.read_csv("results/lambda_results/SGRA_full/abs_err.dat")
#abserr = [float(x) for x in abserr]
#helpers.plot1D(lambdas, abserr, "$\lambda$", "Max Abs Err")

#errstd = helpers.read_csv("results/lambda_results/SGRA_full/std.dat")
#errstd = [float(x) for x in errstd]
#helpers.plot1D(lambdas, errstd, "$\lambda$", "Err $\sigma$")
```
