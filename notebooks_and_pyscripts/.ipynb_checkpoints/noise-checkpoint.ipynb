# Results for all partitions

### Visibility noise with estimates

```python
%%capture
logging.getLogger().setLevel(logging.CRITICAL)

long_names = ["SGRA_long_baselines_15", "SGRA_long_baselines_30", "SGRA_long_baselines_50", \
          "SGRB_long_baselines_15", "SGRB_long_baselines_30", "SGRB_long_baselines_50", \
          "SGRC_long_baselines_15", "SGRC_long_baselines_30", "SGRC_long_baselines_50"]

short_names = ["SGRA_small_baselines_25", "SGRA_small_baselines_40", "SGRA_small_baselines_60", \
          "SGRB_small_baselines_25", "SGRB_small_baselines_40", "SGRB_small_baselines_60", \
          "SGRC_small_baselines_25", "SGRC_small_baselines_40", "SGRC_small_baselines_60"]

long_sizes = [25, 25, 25, 40, 40, 40, 60, 60, 60]

measured_vis = ["measured $\sigma^2$"]
estimated_vis = ["estimated $\sigma^2$"]
measured_recon = ["measured $\eta^2$"]
estimated_recon = ["estimated $\eta^2$"]

def compute_windowed_var(image, window):
    estimated_variance = numpy.zeros(image.shape)

    #convolving initial signal with a variance estimation kernel with a size 2xwindow_hsize+1
    for y in range(image.shape[1]):
        for x in range(image.shape[0]):
            start_x = max(x - window, 0)
            end_x = min(x + window + 1, image.shape[0])
            start_y = max(y - window, 0)
            end_y = min(y + window + 1, image.shape[1])

            estimated_variance[x, y] = numpy.var(image[start_x:end_x, start_y:end_y])
            
    return estimated_variance
    
for i, long_name in enumerate(long_names):
    img_measured = helpers.readFits("../results/noise_results/" + long_name + "_variance_pixel.fits")
    measured_vis.append(numpy.mean(img_measured.flatten()))
    
    [vl] = create_visibility_from_ms('../data4/' + long_name + '.ms')
    dirty, psf = dirty_psf_from_visibilities(vl,cell_size,npix=512)
    windowed_var = compute_windowed_var(dirty.pixels.data[0,0,:,:], 2)
    
    ev = numpy.mean(windowed_var.flatten())
    estimated_vis.append(ev)
    
    #factor = long_sizes[i] / (60*60*1000)
    factor = 1 / 1000
    estimated_recon.append(ev * factor)
    
for short_name in short_names:
    varimg = helpers.readFits("../results/noise_results/" + wavelet_type + "/" + \
                              short_name + "_reconstruction_variance_pixel.fits")
    measured_recon.append(numpy.mean(varimg.flatten()))
    
    [vs] = create_visibility_from_ms('../data4/' + short_name + '.ms')
    dirty, psf = dirty_psf_from_visibilities(vs,cell_size,npix=512)
    windowed_var = compute_windowed_var(dirty.pixels.data[0,0,:,:], 2)
    #ev = numpy.mean(windowed_var.flatten())
    
    #estimated_recon.append(ev)
    
logging.getLogger().setLevel(logging.NOTSET)
```

## For parallelized method, do both with frequencies inverted and different datasets with different observational parameters

```python
dataset_names = ["SGRA", "SGRB", "SGRC"]
gt_names = ["SGRA", "SGRB", "SGRC"]

long_cuts = ["_long_baselines_15", "_long_baselines_30", "_long_baselines_50", \
          "_long_baselines_15", "_long_baselines_30", "_long_baselines_50", \
          "_long_baselines_15", "_long_baselines_30", "_long_baselines_50"]

short_cuts = ["_short_baselines_25", "_short_baselines_40", "_short_baselines_60", \
          "_short_baselines_25", "_short_baselines_40", "_short_baselines_60", \
          "_short_baselines_25", "_short_baselines_40", "_short_baselines_60"]
```

### For $\text{V}_\mathcal{L}$ node

```python
%%capture

logging.getLogger().setLevel(logging.CRITICAL)

n = 50
noise_perc = 2

for i, dataset in enumerate(dataset_names):
    [vt] = create_visibility_from_ms("../data6/" + dataset + ".ms")
    vt, full_cellsize, im = visibilities_from_image(vt,'../data/' + gt_names[i] + '_full_gt.fits',return_cellsize=True,return_image=True,scale_factor=1.8)
    stdv_real = numpy.std(numpy.array(vt.vis.real).flatten())
    stdv_imag = numpy.std(numpy.array(vt.vis.imag).flatten())
    
    for short_cut in short_cuts:
        [vs] = create_visibility_from_ms('../data6/' + dataset + short_cut + '.ms') 
        
        vs, cell_size, im = visibilities_from_image(vs,'../data/' + gt_names[i] + '_full_gt.fits',return_cellsize=True,return_image=True, \
                                                    scale_factor=1.8, override_cellsize=True, ocellsize=full_cellsize)

        single_images = numpy.zeros(512 * 512 * n)
        single_images = numpy.reshape(single_images, (n, 512 * 512))

        for j in range(n):
            viss = helpers.addNoiseToVis(vs, noise_perc, real_deviation=stdv_real, imag_deviation=stdv_imag)
            imgs, psfs = dirty_psf_from_visibilities(viss,cell_size,npix=512)

            single_images[j] = img1.pixels.data.flatten()

        varimgs = numpy.var(single_images, axis=0)
        varimgs = numpy.reshape(varimgs, (512, 512))

        helpers.write_nparr_to_fits(varimgs, "../results/noise_results/parallel/vl/" + dataset + short_cut + "_variance_pixel.fits")

logging.getLogger().setLevel(logging.NOTSET)
```

```python
%%capture

logging.getLogger().setLevel(logging.CRITICAL)

n = 50
noise_perc = 2
num_maj_cycles = 1
fista_iter = 100

tmp_output_name = "tmp_out.fits"
tmp_dirty_name = "tmp_dirty.fits"
tmp_psf_name = "tmp_psf.fits"

for i, dataset in enumerate(dataset_names):
    [vt] = create_visibility_from_ms("../data6/" + dataset + ".ms")
    vt, full_cellsize, im = visibilities_from_image(vt,'../data/' + gt_names[i] + '_full_gt.fits',return_cellsize=True,return_image=True,scale_factor=1.8)
    stdv_real = numpy.std(numpy.array(vt.vis.real).flatten())
    stdv_imag = numpy.std(numpy.array(vt.vis.imag).flatten())
    
    for long_cut in long_cuts:
        [vl] = create_visibility_from_ms('../data6/' + dataset + long_cut + '.ms') 
        
        vl, cell_size, im = visibilities_from_image(vl,'../data/' + gt_names[i] + '_full_gt.fits',return_cellsize=True,return_image=True, \
                                                    scale_factor=1.8, override_cellsize=True, ocellsize=full_cellsize)
        
        recon_images = numpy.zeros(512 * 512 * n)
        recon_images = numpy.reshape(recon_images, (n, 512 * 512))
        initial_lambda = 0.05

        for j in range(n):
            vis = helpers.addNoiseToVis(vl, 2, real_deviation=stdv_real, imag_deviation=stdv_imag)

            img, psf = dirty_psf_from_visibilities(vis,cell_size,npix=512)
            curr_dirty = img.pixels.data
            while len(curr_dirty.shape) > 2:
                curr_dirty = curr_dirty[0]

            curr_psf = psf.pixels.data
            while len(curr_psf.shape) > 2:
                curr_psf = curr_psf[0]

            final_deconvolved = numpy.zeros(curr_dirty.shape)

            for maj_iteration in range(0, num_maj_cycles):
                #deconv
                dirty_norm = numpy.linalg.norm(curr_dirty)
                helpers.write_nparr_to_fits(curr_dirty, tmp_dirty_name)
                helpers.write_nparr_to_fits(curr_psf, tmp_psf_name)
                curr_lambda = initial_lambda * (2 ** maj_iteration) * dirty_norm
                os.system("julia ../julia_rascil_scripts/make_fullres.jl " + str(curr_lambda) + " " + \
                          tmp_psf_name + " " + tmp_dirty_name + " " + str(wavelet_type_idx) + " " + str(fista_iter) + " " + tmp_output_name)
                deconvolved = helpers.readFits(tmp_output_name)

                final_deconvolved += deconvolved

                #de/gridding
                im.pixels.data = [[final_deconvolved]]
                vdec = vl.copy(deep=True)
                vdec = predict_ng(vdec,im,context='2d')
                vres = helpers.subtractVis(vis, vdec)
                model = create_image_from_visibility(vres,cellsize=full_cellsize,npixel=512)
                curr_dirty, sumwt = invert_ng(vres, model, context='2d')
                curr_dirty = curr_dirty.pixels.data

                while len(curr_dirty.shape) > 2:
                    curr_dirty = curr_dirty[0]

            recon_images[j] = final_deconvolved.flatten()

        varimg = numpy.var(recon_images, 0)
        varimg = numpy.reshape(varimg, (512, 512))

        helpers.write_nparr_to_fits(varimg, "../results/noise_results/parallel/vl/" + dataset + long_cut + "_reconstruction_variance_pixel.fits")

logging.getLogger().setLevel(logging.NOTSET)
```

```python
%%capture
logging.getLogger().setLevel(logging.CRITICAL)

measured_vis = ["measured $\sigma^2$"]
estimated_vis = ["estimated $\sigma^2$"]
measured_recon = ["measured $\eta^2$"]
estimated_recon = ["estimated $\eta^2$"]

for dataset in dataset_names:
    for short_cut in short_cuts:
        img_measured = helpers.readFits("../results/noise_results/parallel/vl/" + dataset + short_cut + "_variance_pixel.fits")
        measured_vis.append(numpy.mean(img_measured.flatten()))

        [vs] = create_visibility_from_ms('../data6/' + dataset + short_cut + '.ms')
        dirty, psf = dirty_psf_from_visibilities(vs,cell_size,npix=512)
        windowed_var = helpers.compute_windowed_var(dirty.pixels.data[0,0,:,:], 2)

        ev = numpy.mean(windowed_var.flatten())
        estimated_vis.append(ev)

        factor = 1 / 1000
        estimated_recon.append(ev * factor)

for dataset in dataset_names:
    for long_cut in long_cuts:
        varimg = helpers.readFits("../results/noise_results/parallel/vl/" + \
                                dataset + long_cut +  + "_reconstruction_variance_pixel.fits")
        measured_recon.append(numpy.mean(varimg.flatten()))
    
logging.getLogger().setLevel(logging.NOTSET)
```
